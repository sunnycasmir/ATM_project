Display the menu ("Check Balance" , "Deposit", "Withdraw", "EXIT")
Validate the customer pin. If the customer pin is correct, then it should display "Welcome! and display the menu, otherwise it can allow customer to try for a maximum of 3 times after which the customer card would be blocked.
Check the current balance of the customer (You can assume customer has zero balance at first)
Deposit money to your account.  Display the message "Successful deposit!" and show current balance
Withdraw money from the account and display balance amount. Display the message "Successful withdrawal!" if amount to withdraw is possible otherwise "Insufficient balance!".
The program should exit when customer presses the EXIT button
 

```python
class Bank:
    
#initialize the class with a special with attributes included 

    def __init__(self):
        self.balance = 0
        self.attempts = 0
        self.blocked = False
        self.customer_pin = "1182"
        self.heighest_attempt = 3

#creat a function to check customers pin validation
    def validatePin(self):
        if self.attempts ==  self.heighest_attempt:
            self.blocked = True
            print(f" You card is now blocked,maximum attempts exceeded.")
            return False
        else:
            pin = input("Please enter your pin: ")
            if pin == self.customer_pin:
                print("Welcome!")
                return True
            else:
                self.attempts += 1
                print(f"try again,Incorrect pin")
                return self.validatePin()

 #create a function to display menu
    def displayMenu(self):
        print("Please select an option:")
        print("1. Check Balance")
        print("2. Deposit")
        print("3. Withdraw")
        print("4. Exit")

# create a function for checkking balance
    def checkBalance(self):
        print("Your current balance is: $" + str(self.balance))

# create a function for depositing money
    def deposit(self):
        amount = float(input("Please enter the amount you would like to deposit: $"))
        self.balance += amount
        print("Successful deposit!")
        self.checkBalance()

 # create a function for withdrawing money
    def withdraw(self):
        amount = float(input("Please enter the amount you would like to withdraw: $"))
        if self.balance >= amount:
            self.balance -= amount
            print("Successful withdrawal!")
            self.checkBalance()
        else:
            print("Insufficient balance!")
            
# create a function to select options and exit
    def run_account(self):
        if self.validatePin():
            while not self.blocked:
                self.displayMenu()
                choice = input()
                if choice == "1":
                    self.checkBalance()
                elif choice == "2":
                    self.deposit()
                elif choice == "3":
                    self.withdraw()
                elif choice == "4":
                    print("Thank you for using our ATM!")
                    self.blocked = True
                else:
                    print("Invalid choice. Please try again.")
```

```python
account = Bank()
account.run_account()
```
